```python
def find(number):
    if number == 1:
        return 1
    elif number ==0:
        return 1
    return number * find(number - 1)

print(find(3))
```

```python
jobs = info.find_all(class_='column is-half')
jobs
```

```python
for jobNum in range(len(jobs)):
    print(f'Post number {jobNum+1}')
    jobtitle = jobs[jobNum].find(class_='title is-5')
    jobdesc = jobs[jobNum].find(class_='subtitle is-6 company')
    location = jobs[jobNum].find(class_='location')
    date = jobs[jobNum].find(class_= 'is-small has-text-grey')
    print(f'Job Title is  f{jobtitle.text.strip()}')
    print(f'Job description is  {jobdesc.text.strip()}')
    print(f'Job location is  {location.text.strip()}')
    print(f'Job date is  {date.text.strip()}')
    print("-"*50)
    print()
```
